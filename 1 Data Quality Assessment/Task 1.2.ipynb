```python
# check for null values
df2.isnull().sum()
```

```python
# Dropping the unnamed columns with unknown datapoints
del df2['Unnamed: 16']
del df2['Unnamed: 17']
del df2['Unnamed: 18']
del df2['Unnamed: 19']
del df2['Unnamed: 20']
```

```python
#sort by first name then fill the missing last name.
df2['last_name'] = df2.groupby('first_name').last_name.bfill().ffill()
```

```python
# check the entered data
df2['gender'].value_counts()
```

```python
# check the entered data
df2['state'].value_counts()
```

```python
# check the entered data
df2['country'].value_counts()
```

```python
# check the entered data
df2['address'].nunique()
```

```python
# make sure all dates are in the correct format
df2['DOB'] = df2['DOB'].dt.date
df2['DOB']
```

```python
# fill the missing date of birth with a random choice of date from the existing datapoints
df2['DOB'].fillna(lambda x: np.random.choice(df2['DOB']), inplace=True)
```

```python
# fill the missing datapoint using the backward filling method
df2['job_title']= df2['job_title'].fillna(method='bfill')
```

```python
# check the frequency of job industries
df2['job_industry_category'].value_counts()
```

```python
# fill the missing datapoint using the forward filling method
df2['job_industry_category']= df2['job_industry_category'].fillna(method='ffill')
```

```python
# make sure that the column names are in the correct format
df2.columns = map(str.lower, df2.columns)
df2.columns = map(str.strip, df2.columns)
```

```python
# round the values to the correct format
df2['value'] = df2['value'].round(3)
df2
```
